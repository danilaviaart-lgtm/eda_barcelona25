```python
import pandas as pd
import numpy as np
```

```python
df_barcelona = pd.read_csv("src/sexo_edad.csv")
df_barcelona
```

```python
punts = pd.read_csv("src/puntos_carga_barcelona_v2.csv")

punts['cpostal'] = punts['direccion'].str[-4:]
punts['cpostal'] = punts['cpostal'].replace('dral', '8002')

```

```python
punts['cpostal'] = punts['cpostal'].replace('trum', '8039')
```

```python
postals = pd.read_csv("src/result/pcode.csv")
postals

postals = postals.rename(columns={"Códigos Postales" : "cpostal"})
punts["cpostal"] = punts["cpostal"].astype(int)
```

```python
endolls = pd.merge(punts, postals, on='cpostal')
endolls.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
endolls.head()
```

```python
columnas_sumar = ['type2', 'schuko', 'ccs2', 'chademo']
endolls['total_punts'] = endolls[columnas_sumar].sum(axis=1)
```

```python
endolls['total_punts'] = endolls['total_punts'].replace(0, 1)
endolls.head()
```

```python
endolls = endolls.rename(columns={"Distrito" : "Nom_Districte"})
```

```python
endolls.to_csv("src/result/endolls.csv")
```
