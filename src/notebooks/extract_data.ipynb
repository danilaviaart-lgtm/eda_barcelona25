```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# This line handles the version matching automatically
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get("https://www.electromaps.com/es/puntos-carga/espana/barcelona/barcelona")
```

```python
puntos_carga  = {
    "nombre":[],
    "tipo":[],
    "direccion":[],
    "type2":[],
    "schuko":[],
    "ccs2":[],
    "chademo":[],
}
```

```python
while True:
    try:
        boton_mas = driver.find_element(By.ID, "loadMoreStations")
        boton_mas.click()
        time.sleep(2)  
    except (NoSuchElementException, ElementClickInterceptedException):
        print("No hay más botones.")
        break
```

```python
def extraer_texto(contexto, xpath, default="N/A"):
    try:
        return contexto.find_element(By.XPATH, xpath).text
    except NoSuchElementException:
        return default

lista_estaciones = driver.find_elements(By.CSS_SELECTOR, "#stations > a")
for nodo in lista_estaciones:
    nombre = extraer_texto(nodo, ".//div[2]/h3", "Sin nombre")
    tipo = extraer_texto(nodo, ".//div[1]/div[1]/div[1]//div[2]/div", "Sin tipo")
    direccion = extraer_texto(nodo, ".//div[3]/div", "Sin dirección")
    
    type2 = extraer_texto(nodo, ".//div[4]/div[@title='TYPE 2']/div[2]/div[2]", "0")
    schuko = extraer_texto(nodo, ".//div[4]/div[@title='Schuko (EU Plug)']/div[2]/div[2]", "0")
    ccs2 = extraer_texto(nodo, ".//div[4]/div[@title='CCS2']/div[2]/div[2]", "0")
    chademo = extraer_texto(nodo, ".//div[4]/div[@title='CHAdeMO']/div[2]/div[2]", "0")
    
    puntos_carga["nombre"].append(nombre)
    puntos_carga["tipo"].append(tipo)
    puntos_carga["direccion"].append(direccion)
    puntos_carga["type2"].append(type2)
    puntos_carga["schuko"].append(schuko)
    puntos_carga["ccs2"].append(ccs2)
    puntos_carga["chademo"].append(chademo)
```

```python
guardar = pd.DataFrame(puntos_carga)
guardar.to_csv("puntos_carga_barcelona_v2.csv")
```
