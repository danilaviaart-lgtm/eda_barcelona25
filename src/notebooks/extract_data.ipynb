```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# This line handles the version matching automatically
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get("https://www.electromaps.com/es/puntos-carga/espana/barcelona/barcelona")
```

```python
puntos_carga  = {
    "nombre":[],
    "tipo":[],
    "direccion":[],
    "type2":[],
    "schuko":[],
    "ccs2":[],
    "chademo":[],
}
```

```python
while True:
    try:
        boton_mas = driver.find_element(By.ID, "loadMoreStations")
        boton_mas.click()
        time.sleep(2)  
    except (NoSuchElementException, ElementClickInterceptedException):
        print("No hay más botones.")
        break

```

```python
def extraer_texto(contexto, xpath, default="N/A"):
    try:
        return contexto.find_element(By.XPATH, xpath).text
    except NoSuchElementException:
        return default

lista_estaciones = driver.find_elements(By.CSS_SELECTOR, "#stations > a")
for nodo in lista_estaciones:
    nombre = extraer_texto(nodo, ".//div[2]/h3", "Sin nombre")
    tipo = extraer_texto(nodo, ".//div[1]/div[1]/div[1]//div[2]/div", "Sin tipo")
    direccion = extraer_texto(nodo, ".//div[3]/div", "Sin dirección")
    
    type2 = extraer_texto(nodo, ".//div[4]/div[@title='TYPE 2']/div[2]/div[2]", "0")
    schuko = extraer_texto(nodo, ".//div[4]/div[@title='Schuko (EU Plug)']/div[2]/div[2]", "0")
    ccs2 = extraer_texto(nodo, ".//div[4]/div[@title='CCS2']/div[2]/div[2]", "0")
    chademo = extraer_texto(nodo, ".//div[4]/div[@title='CHAdeMO']/div[2]/div[2]", "0")
    
    puntos_carga["nombre"].append(nombre)
    puntos_carga["tipo"].append(tipo)
    puntos_carga["direccion"].append(direccion)
    puntos_carga["type2"].append(type2)
    puntos_carga["schuko"].append(schuko)
    puntos_carga["ccs2"].append(ccs2)
    puntos_carga["chademo"].append(chademo)

        
        
```

```python
nombre_e = driver.find_element(By.CLASS_NAME, "title-charge-point").text
tipo_e = driver.find_element(By.CLASS_NAME, "chargepoints-stations-visual-text").text
direccion = driver.find_element(By.CLASS_NAME, "char-aside-info-text").text

print(nombre_e)
print(tipo_e)
print(direccion)

items = driver.find_elements(By.CLASS_NAME, "char-connectors-item")

for item in items:
    if item.find_element(By.CLASS_NAME, "char-connectors-title").text == "Schuko (EU Plug)":
        cargadores["schuko"].append(item.find_element(By.CLASS_NAME, "char-lenght").text)
        cargadores["schuko-kw"].append(item.find_element(By.CLASS_NAME, "char-kw").text)
    else:
        cargadores["schuko"].append(0)
        cargadores["schuko-kw"].append(0)
    if item.find_element(By.CLASS_NAME, "char-connectors-title").text == "CCS2":
        cargadores['ccs2'].append(item.find_element(By.CLASS_NAME, "char-lenght").text)
        cargadores['ccs2-kw'].append(item.find_element(By.CLASS_NAME, "char-kw").text)
    else:
        cargadores['ccs2'].append(0)
        cargadores['ccs2-kw'].append(0)
    if item.find_element(By.CLASS_NAME, "char-connectors-title").text == "TYPE 2":
        cargadores['type2'].append(item.find_element(By.CLASS_NAME, "char-lenght").text)
        cargadores['type2-kw'].append(item.find_element(By.CLASS_NAME, "char-kw").text)
    else:
        cargadores['type2'].append(0)
        cargadores['type2-kw'].append(0)
    if item.find_element(By.CLASS_NAME, "char-connectors-title").text == "CHAdeMO":
        cargadores['chademo'].append(item.find_element(By.CLASS_NAME, "char-lenght").text)
        cargadores['chademo-kw'].append(item.find_element(By.CLASS_NAME, "char-kw").text)
    else:
        cargadores['chademo'].append(0)
        cargadores['chademo-kw'].append(0)
        
```
