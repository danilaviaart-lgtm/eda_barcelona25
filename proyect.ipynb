### ADN BARCELONA 2025

## 1. Introducción y Contexto
El objetivo de este análisis es determinar si la **densidad de población**, el **nivel de renta** y las **características del parque móvil** en los distritos de Barcelona están relacionados y si es posible una transición ecológica.

* **Origen de los datos:** Registros públicos del Ayuntamiento de Barcelona y WebScrapping de la web Electromaps.
* **Alcance:** El estudio se centra exclusivamente en los datos de cargadores públicos disponibles.

---

## 2. Análisis Demográfico y Económico
La ciudad muestra una profunda polarización, con una brecha de renta de casi el 100% entre el distrito más rico y el más pobre.

* **Sarrià - St. Gervasi:** Población baja, pero la renta más alta de la ciudad.
* **L’Eixample:** El motor económico; es el distrito más poblado y mantiene un nivel de renta alto.
* **Nou Barris y Ciutat Vella:** Concentran los niveles de renta más bajos, evidenciando una brecha de vulnerabilidad social.

## 3. Radiografía del Parque Móvil
En Barcelona, **1 de cada 2 habitantes** dispone de vehículo propio.

* **Líderes en motorización:** Sarrià y Les Corts se sitúan por encima de la media. En Sarrià-St. Gervasi hay casi un vehículo por persona.
* **Zonas Peatonales:** Ciutat Vella presenta el parque móvil más reducido debido a las restricciones de movilidad y facilidad de transporte público.
* **Gràcia:** Muestra una baja tasa de vehículos, probablemente por su estructura de calles estrechas y fomento del transporte peatonal.

## 4. El Reto del Envejecimiento (Antigüedad)
Casi la mitad de los vehículos de la ciudad (**48.5%**) tienen más de 10 años.

* **Reto ambiental:** Estos vehículos complican los objetivos de las Zonas de Bajas Emisiones (ZBE) al ser más contaminantes.
* **Desigualdad en la renovación:** Los distritos con mayor poder adquisitivo (Sarrià-Sant Gervasi, Les Corts) tienen coches más nuevos. En cambio, en Nou Barris y Horta-Guinardó, la baja renta dificulta la renovación tecnológica.

## 5. Tipos de Combustión e Infraestructura
Barcelona sigue siendo una ciudad mayoritariamente "térmica":
* **Combustión tradicional:** 723,528 vehículos (Gasolina y Diésel).
* **Vehículos ECO:** 75,410 vehículos (Híbridos y Eléctricos), apenas una décima parte del total.

### Infraestructura de Carga
* **Ubicación:** El 65.8% de los cargadores están en superficie y el 23.8% en parkings públicos.
* **Carencias técnicas:** La red está diseñada para estancias medias (Type 2 y Schuko). Falta infraestructura de carga ultra-rápida (CCS2).
* **Brecha territorial:** Los cargadores se concentran en el centro y zonas ricas, dejando rezagados a barrios periféricos como Nou Barris.

## 6. Conclusiones y Estrategia

### Conclusiones Principales
1.  **División por renta:** La movilidad privada y la calidad de la flota están ligadas al nivel económico.
2.  **Obsolescencia:** El elevado porcentaje de coches antiguos (>10 años) es el mayor freno para los objetivos ambientales.
3.  **Desigualdad ecológica:** La transición energética actual corre el riesgo de ser exclusiva de las rentas altas si no se invierte en la periferia.

### Estrategia Propuesta
* **Incentivos:** Programas de ayudas para renovar vehículos en distritos de renta baja (Nou Barris, Ciutat Vella).
* **Reequilibrio:** Desplegar cargadores públicos de forma prioritaria en barrios periféricos.
* **Modernización:** Fomentar la carga ultra-rápida en estaciones de servicio y centros comerciales.
